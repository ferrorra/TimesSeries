#### Idee : 
 Je pense que hours.csv est inclu dans day.csv donc n va verifier ceci ensuite on fait un merge sur les colonnes communes et on rajoute l'attribut hr

```python
df_day['dteday'] = pd.to_datetime(df_day['dteday'])
df_hour['dteday'] = pd.to_datetime(df_hour['dteday'])
```

```python
common_cols = [col for col in df_day.columns if col in df_hour.columns]
```

```python
contained = df_day[common_cols].isin(df_hour[common_cols]).any()
```

```python
false_columns = contained[~contained].index.tolist()
false_columns 
```

Je vais donc bien analyser ces variables, si c'est des variables numeriques, il se peut que c'est des moyennes/ du jour (calculee a partir des valeur de chaque heure) ou bien des sommes des heures dans hour.csv mises dans day.csv
- **atemp**: Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale) (moyenne)
- **windspeed**: Normalized wind speed. The values are divided to 67 (max) (moyenne)
- **casual**: count of casual users (somme)
- **registered**: count of registered users (somme)
- **cnt**: count of total rental bikes including both casual and registered (somme des sommes )



```python
df_hour_grouped = df_hour.groupby(df_hour['dteday'].dt.date).agg({'atemp': 'mean', 'windspeed': 'mean', 'casual': 'sum', 'registered': 'sum', 'cnt': 'sum'})
```

```python
for col in ['atemp', 'windspeed', 'casual', 'registered', 'cnt']:
    print(f"{col}: {np.allclose(df_day.set_index('dteday')[col], df_hour_grouped[col], equal_nan=True)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df_day = pd.read_csv('bike+sharing+dataset/day.csv')
df_hour = pd.read_csv('bike+sharing+dataset/hour.csv')


plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='temp', data=df_day)
plt.title('Temperature across seasons')
plt.xlabel('Season')
plt.ylabel('Normalized Temperature')
plt.xticks([0, 1, 2, 3], ['Spring', 'Summer', 'Fall', 'Winter'])
plt.show()

mean_temp = df_day['temp'].mean()
median_temp = df_day['temp'].median()
print(f"Mean temperature: {mean_temp}")
print(f"Median temperature: {median_temp}")
```
